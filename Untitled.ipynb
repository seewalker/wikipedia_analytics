```python
from numpy.random import multivariate_normal
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.
```

```python
mean = [0.1,1.0]
covar = [[2,3],[-3,15]]
result = multivariate_normal(mean,covar,4000)
xs, ys = result[:,0], result[:,1]
#plt.scatter(xs,ys, s=4 )
plt.scatter(xs,ys)
```

```python
sns.pointplot(xs,ys)
```

```python
sns.set_style("ticks")
sns.pointplot(xs[0:40],ys[0:40], markers='x')
```

```python
sns.countplot([1,1,1,2,2,3])
plt.hist([0,0,1])
```

```python
sns.jointplot(xs,ys)
```
